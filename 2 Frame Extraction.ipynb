```python
# Extract the corresponding .csv record given a MP4 path e.g. 'MELD.Raw/train_splits/dia632_utt5.mp4'
def get_csv_record(filename):
    """
    Given a filename (e.g., '/content/drive/MyDrive/CS5344 Project/Data and Codes/dev_videos/dia15_utt4.mp4'), retrieves the corresponding record from the corresponding folder.

    :param filename: The video filename with the folder path (e.g., '/content/drive/MyDrive/CS5344 Project/Data and Codes/dev_videos/dia15_utt4.mp4').
    :return: The corresponding record from the dataframe or an error message.
    """
    # Extract the folder and video filename from the full path
    folder = filename.split('/')[6]
    dialogue_id = int(filename.split('dia')[1].split('_')[0])
    utterance_id = int(filename.split('utt')[1].split('.mp4')[0])

    # Filter the dataframe for the corresponding record
    if folder == 'train_videos':
        record = train_df[(train_df['Dialogue_ID'] == dialogue_id) & (train_df['Utterance_ID'] == utterance_id)]
    elif folder == 'dev_videos':
        record = dev_df[(dev_df['Dialogue_ID'] == dialogue_id) & (dev_df['Utterance_ID'] == utterance_id)]
    elif folder == 'test_videos':
        record = test_df[(test_df['Dialogue_ID'] == dialogue_id) & (test_df['Utterance_ID'] == utterance_id)]

    # Return the record if found
    if not record.empty:
        print(filename)
        display(record)
        return record
    else:
        return f"No record found for file: {filename}"
```

```python
# Extracting frames from a video - for preview
def extract_frame(video_path, frame_num=0):
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)  # Move to the desired frame
    ret, frame = cap.read()
    cap.release()

    if ret:
        frame = cv2.resize(frame, (1280, 720))
        cv2_imshow(frame)  # Show the frame in Colab
    else:
        print("Error: Could not read frame.")

# Test on the first video
extract_frame(video_path)
```

### 4. Extract the middle frame for each video file

```python
import numpy as np
from glob import glob
```

```python
# Directories containing MP4 files
data_dirs = {
    "train": "/content/drive/MyDrive/CS5344 Project/Data and Codes/train_videos/",
    "dev": "/content/drive/MyDrive/CS5344 Project/Data and Codes/dev_videos",
    "test": "/content/drive/MyDrive/CS5344 Project/Data and Codes/test_videos"
}

# Output directories for extracted frames
output_dirs = {
    "train": "/content/drive/MyDrive/CS5344 Project/Data and Codes/train_frames/",
    "dev": "/content/drive/MyDrive/CS5344 Project/Data and Codes/dev_frames/",
    "test": "/content/drive/MyDrive/CS5344 Project/Data and Codes/test_frames/"
}

# Create output directories if they do not exist
for key, out_dir in output_dirs.items():
    os.makedirs(out_dir, exist_ok=True)

# Function to extract a single frame from a video
def extract_one_frame(video_path, save_path):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if frame_count == 0:
        print(f"Skipping {video_path}, no frames found.")
        return None

    middle_frame_idx = frame_count // 2  # Choose middle frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, middle_frame_idx)  # Jump to middle frame

    success, frame = cap.read()
    cap.release()

    if success:
        cv2.imwrite(save_path, frame)
        return frame

    return None

# Process all MP4 files in respective directories
# Save the 1 middle frame from each MP4 file in the corresponding directory
for split, video_dir in data_dirs.items():
    output_dir = output_dirs[split]
    video_files = glob(os.path.join(video_dir, "*.mp4"))

    for video_file in video_files:
        video_name = os.path.splitext(os.path.basename(video_file))[0]
        frame_path = os.path.join(output_dir, f"{video_name}.jpg")

        # Extract and save frame
        frame = extract_one_frame(video_file, frame_path)

print("Frame extraction completed.")
```
